```python
from typing import Tuple, List, Type
from urllib.parse import urlparse

URL = Type[str]
import requests as r
from bs4 import BeautifulSoup, Tag
```

```python
url = 'https://med.emory.edu/departments/radiation-oncology/research-laboratories/deformable-image-registration/downloads-and-reference-data/4dct.html'
url_info = urlparse(url)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
res = r.get(url, headers=headers)
html = res.content
soup = BeautifulSoup(html, 'html.parser')
```

```python
visited = set()
for a_element in soup.find_all('a'):
    a_element: Tag
    link = a_element.attrs['href']
    clear_link = link.strip('/')

    if link.startswith('http'):
        link_info = urlparse(link)
        if link_info.hostname == url_info.hostname:
            if clear_link != url.rstrip('/'):
                visited.add(clear_link)
        else:
            print(link_info)
    elif link.startswith('/'):
        clear_url = url.rstrip('/')
        target_url = clear_url + '/' + clear_link
        visited.add(target_url)
    else:
        visited.add(link)
```

```python
for link in visited:
    if not link.endswith('html'):
        print(link)
```
